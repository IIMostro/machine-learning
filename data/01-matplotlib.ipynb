```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
```

```python
x
```

```python
sinx = np.sin(x)
```

```python
cosx = np.cos(x)
```

```python
plt.plot(x, sinx)
plt.show()
```

```python
plt.plot(x, sinx, color='red')
plt.plot(x, cosx, color='blue', linestyle="--")
plt.show()
```

```python
plt.plot(x, sinx, color='red')
plt.plot(x, cosx, color='blue', linestyle="--")
plt.axis([-5, 15, -2, 2])
plt.show()
```

```python
plt.plot(x, sinx, color='red')
plt.plot(x, cosx, color='blue', linestyle="--")
plt.xlabel("x lable")
plt.ylabel("y lable")
plt.show()
```

```python
plt.plot(x, sinx, color='red', label="sinx")
plt.plot(x, cosx, color='blue', linestyle="--", label="cosx")
plt.xlabel("x lable")
plt.ylabel("y lable")
plt.legend()
plt.show()
```

```python
plt.plot(x, sinx, color='red', label="sinx")
plt.plot(x, cosx, color='blue', linestyle="--", label="cosx")
plt.xlabel("x lable")
plt.ylabel("y lable")
plt.title("Welcome to the ML World!")
plt.legend()
plt.show()
```

### Scatter Plot

```python
plt.scatter(x, sinx)
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)

plt.scatter(x, y)
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)

plt.scatter(x, y, alpha=0.5)
plt.show()
```
